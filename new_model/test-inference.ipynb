```python
from ultralytics import YOLO
import cv2

# Carregar o modelo treinado
model = YOLO("best.pt")

# Executar a predição no vídeo
results = model.predict(
    source="images/video_teste.mp4", 
    conf=0.50,
    stream=True,  # Ativa o modo de processamento quadro a quadro
    save=True,    # Salva o vídeo com as detecções
    show=True,    # Mostra a janela de visualização
    device=0      # 0 para GPU; "cpu" para CPU
)

for result in results:
    pass

print("Processamento do vídeo concluído.")
print("O vídeo salvo está na pasta: runs/detect/predict/")

cv2.destroyAllWindows()
```
